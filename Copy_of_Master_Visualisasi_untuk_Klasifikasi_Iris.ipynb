```python
df['Species']=df['Species'].astype(int)
```

```python
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'], c=df['Species'], cmap=plt.cm.get_cmap('viridis', 3))
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.colorbar()
plt.show()
```

```python
df=df.drop('Id',1)
df1=df.drop('PetalLengthCm', 1)
df1=df1.drop('PetalWidthCm',1)
df1
```

```python
x=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
```

```python
le = preprocessing.LabelEncoder()
le.fit(y)
list(le.classes_)
y=le.transform(y)
print(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state=2)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
clf=tree.DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
pred=clf.predict(x_test)
print(pred)
```

```python
print('Akurasi Klasifikasi: ',metrics.accuracy_score(pred,y_test)*100,'%')
fpr, tpr, thresholds = metrics.roc_curve(y_test, pred, pos_label=2)
print('Nilai AUC Klasifikasi: ', metrics.auc(fpr, tpr))
print('Nilai TPR (True Positive Rate) Klasifikasi: ', tpr)
print('Nilai FPR (False Positive Rate) Klasifikasi: ', fpr)
```
