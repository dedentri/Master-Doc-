2. Menggunakan Elbow method tentukan jumlah cluster yang optimal untuk dataset ini dengan variabel: 'Attack', 'Defense'

image.png

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WSCC')

plt.show()
```

4. Visualisasikan  'Attack' dan 'Defense' plot ke dalam 2D setelah dicluster. Gambar di bawah ini adalah 3 cluster. ingat kalian adalah 5 cluster. Lengkap dengan title, label, legend

image.png


```python
from sklearn.cluster import KMeans
import numpy as np
# k means
kmeans = KMeans(n_clusters=5, random_state=0)
df['cluster'] = kmeans.fit_predict(df[['Attack', 'Defense']])
# get centroids
centroids = kmeans.cluster_centers_
cen_x = [i[0] for i in centroids]
cen_y = [i[1] for i in centroids]
## add to df
df['cen_x'] = df.cluster.map({0:cen_x[0], 1:cen_x[1], 2:cen_x[2], 3:cen_x[3], 4:cen_x[4]})
df['cen_y'] = df.cluster.map({0:cen_y[0], 1:cen_y[1], 2:cen_y[2], 3:cen_y[3], 4:cen_y[4]})
# define and map colors
colors = ['red', 'green', 'blue', 'cyan', 'magenta']
df['c'] = df.cluster.map({0:colors[0], 1:colors[1], 2:colors[2], 3:colors[3], 4:colors[4]})
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.Attack, df.Defense, c=df.c, alpha = 0.6, s=10)
plt.title('Pokemon Stats\n', loc='left', fontsize=22)
plt.xlabel('Attack')
plt.ylabel('Defense')
```

6. Visualisasikan cluster-cluster dengan smooth polygon. Visulaisasi di bawah ini untuk Attack dan Defense. Masukkan variabel 'Speed' yang menjadikan ukuran titik yang berbeda beda

image.png

```python
from scipy.spatial import ConvexHull
import numpy as np
fig, ax = plt.subplots(1, figsize=(8,8))
# plot data
plt.scatter(df.Attack, df.Defense, c=df.c, alpha = 0.6, s=z)
# plot centers
plt.scatter(cen_x, cen_y, marker='^', c='black', s=100)
# draw enclosure
for i in df.cluster.unique():
    points = df[df.cluster == i][['Attack', 'Defense']].values
    # get convex hull
    hull = ConvexHull(points)
    # get x and y coordinates
    # repeat last point to close the polygon
    x_hull = np.append(points[hull.vertices,0],
                       points[hull.vertices,0][0])
    y_hull = np.append(points[hull.vertices,1],
                       points[hull.vertices,1][0])
    # plot shape
    plt.fill(x_hull, y_hull, alpha=0.3, c=colors[i])

plt.xlim(0,250)
plt.ylim(0,250)
```
